```python
import pandas as pd
df = pd.read_csv("data/train.csv")
df.head(5)
```

```python
df[["Sex","Age","SibSp","Parch"]]
```

```python
df_test = pd.read_csv("data/test.csv")
df_test.head()
```

```python
df.loc[df["Age"]<10, "Age"] = 0
df.loc[(df["Age"]>=10) & (df["Age"]<20), "Age"] = 10
df.loc[(df["Age"]>=20) & (df["Age"]<30), "Age"] = 20
df.loc[(df["Age"]>=30) & (df["Age"]<40), "Age"] = 30
df.loc[(df["Age"]>=40) & (df["Age"]<50), "Age"] = 40
df.loc[df["Age"]>=50, "Age"] = 50
df
```

```python
df_test.loc[df_test["Age"]<10, "Age"] = 0
df_test.loc[(df_test["Age"]>=10) & (df_test["Age"]<20), "Age"] = 10
df_test.loc[(df_test["Age"]>=20) & (df_test["Age"]<30), "Age"] = 20
df_test.loc[(df_test["Age"]>=30) & (df_test["Age"]<40), "Age"] = 30
df_test.loc[(df_test["Age"]>=40) & (df_test["Age"]<50), "Age"] = 40
df_test.loc[df_test["Age"]>=50, "Age"] = 50
df_test
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]
df.head()
```

```python
df_test["FamilySize"] = df_test["SibSp"] + df_test["Parch"]
df_test.head()
```

```python
train = df[["Survived", "Sex", "Age", "FamilySize"]]
test = df_test[["Sex", "Age", "FamilySize"]]
train.head()
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
df_test["Fare"] = df_test["Fare"].fillna(df_test["Fare"].mean())
df_test.head()
```

```python
df.loc[df["Name"].str.contains("Mr\."), "Name"] = "Mr"
df.loc[df["Name"].str.contains("Mrs\."), "Name"] = "Mrs"
df.loc[df["Name"].str.contains("Miss\."), "Name"] = "Miss"
```

```python
df["Name"] = df["Name"].map({
    "Mr": 0,
    "Mrs": 1,
    "Miss": 2
})
df["Name"] = df["Name"].fillna(3)
df
```

```python
df["Age"] = df["Age"].fillna(df.groupby("Name")["Age"].transform("mean"))
df
```
