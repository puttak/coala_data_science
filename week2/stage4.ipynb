```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.plot([1,2,3], [90,100,50])
plt.show()
```

```python
plt.plot(["Mat", "Kor", "Eng"], [90, 100, 50])
plt.plot(["Mat", "Kor", "Eng"], [70, 60, 90])
plt.xlabel("Sub")
plt.ylabel("Score")
plt.title("Midterm")
plt.legend(["Choi", "Kim"])
plt.show()
```

```python
x = ["Mat", "Kor", "Eng"]
y = [90, 100, 50]

plt.xlabel("Sub")
plt.ylabel("Score")

plt.bar(x, y, width = 0.3, color = "#666666")
plt.show()
```

```python
plt.plot([1,2,3],[90,100,50], marker="+", markersize = 30
        , linestyle=":")
plt.grid(True)
plt.axis([0,3,50,100])

plt.show()
```

```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")
train.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.relplot(data=train, x="Age", y="Survived")
```

```python
sns.relplot(data=train, x="Age", y="Survived", hue="Sex", aspect=2)
```

```python
sns.catplot(data=train, x="Pclass", y="Fare", aspect=2, 
            jitter=False, hue="Sex")
```

```python
sns.pairplot(data=train, hue="Sex",
            x_vars=["Pclass", "Age", "Fare"],
            y_vars=["Survived", "Age", "Fare"], height=3)
```

```python
sns.catplot(data=train, x="Pclass", y="Fare", aspect=2, 
             hue="Sex", kind="box")
```
